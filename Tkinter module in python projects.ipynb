```python
import tkinter as tk

class PixelArtEditor:
    def __init__(self, root, width, height, pixel_size):
        self.root = root
        self.pixel_size = pixel_size
        self.canvas_width = width * pixel_size
        self.canvas_height = height * pixel_size
        self.canvas = tk.Canvas(root, width=self.canvas_width, height=self.canvas_height, bg='white')
        self.canvas.pack()

        self.grid = [[False for _ in range(width)] for _ in range(height)]
        self.draw_color = 'black'

        self.canvas.bind('<B1-Motion>', self.draw_pixel)
        self.canvas.bind('<ButtonPress-1>', self.draw_pixel)

    def draw_pixel(self, event):
        x, y = event.x // self.pixel_size, event.y // self.pixel_size
        if 0 <= x < len(self.grid[0]) and 0 <= y < len(self.grid):
            self.grid[y][x] = self.draw_color
            x1, y1 = x * self.pixel_size, y * self.pixel_size
            x2, y2 = x1 + self.pixel_size, y1 + self.pixel_size
            self.canvas.create_rectangle(x1, y1, x2, y2, fill=self.draw_color, outline='')

    def set_draw_color(self, color):
        self.draw_color = color

    def clear_canvas(self):
        self.grid = [[False for _ in range(len(self.grid[0]))] for _ in range(len(self.grid))]
        self.canvas.delete('all')

    def save_as_text_file(self, filename):
        with open(filename, 'w') as file:
            for row in self.grid:
                for pixel_color in row:
                    file.write(f"{pixel_color} ")
                file.write("\n")

if __name__ == "__main__":
    WIDTH, HEIGHT = 20, 20  # Change the width and height to adjust the canvas size
    PIXEL_SIZE = 20  # Change the pixel size for smaller/larger pixels

    root = tk.Tk()
    root.title("Pixel Art Editor")

    editor = PixelArtEditor(root, WIDTH, HEIGHT, PIXEL_SIZE)

    colors = ['black', 'red', 'blue', 'green', 'yellow', 'orange', 'purple', 'white', 'pink', 'cyan']
    color_buttons = []
    for color in colors:
        color_button = tk.Button(root, bg=color, width=3, command=lambda c=color: editor.set_draw_color(c))
        color_button.pack(side='left', padx=2)
        color_buttons.append(color_button)

    clear_button = tk.Button(root, text='Clear', command=editor.clear_canvas)
    clear_button.pack()

    save_button = tk.Button(root, text='Save as Text', command=lambda: editor.save_as_text_file('pixel_art.txt'))
    save_button.pack()

    root.mainloop()
```

```python
import tkinter as tk

class PixelArtEditor:
    def __init__(self, root, width, height, pixel_size):
        self.root = root
        self.pixel_size = pixel_size
        self.canvas_width = width * pixel_size
        self.canvas_height = height * pixel_size
        self.canvas = tk.Canvas(root, width=self.canvas_width, height=self.canvas_height, bg='white')
        self.canvas.pack()

        self.grid = [['white' for _ in range(width)] for _ in range(height)]
        self.draw_color = 'black'

        self.canvas.bind('<B1-Motion>', self.draw_pixel)
        self.canvas.bind('<ButtonPress-1>', self.draw_pixel)

    def draw_pixel(self, event):
        x, y = event.x // self.pixel_size, event.y // self.pixel_size
        if 0 <= x < len(self.grid[0]) and 0 <= y < len(self.grid):
            self.grid[y][x] = self.draw_color
            x1, y1 = x * self.pixel_size, y * self.pixel_size
            x2, y2 = x1 + self.pixel_size, y1 + self.pixel_size
            self.canvas.create_rectangle(x1, y1, x2, y2, fill=self.draw_color, outline='')

    def set_draw_color(self, color):
        self.draw_color = color

    def clear_canvas(self):
        self.grid = [['white' for _ in range(len(self.grid[0]))] for _ in range(len(self.grid))]
        self.canvas.delete('all')

    def save_as_text_file(self, filename):
        with open(filename, 'w') as file:
            for row in self.grid:
                file.write(' '.join(row) + '\n')

    def load_from_text_file(self, filename):
        with open(filename, 'r') as file:
            lines = file.readlines()
            self.grid = [[color.strip() for color in line.split()] for line in lines]

        self.canvas.delete('all')
        for y, row in enumerate(self.grid):
            for x, pixel_color in enumerate(row):
                x1, y1 = x * self.pixel_size, y * self.pixel_size
                x2, y2 = x1 + self.pixel_size, y1 + self.pixel_size
                self.canvas.create_rectangle(x1, y1, x2, y2, fill=pixel_color, outline='')

if __name__ == "__main__":
    WIDTH, HEIGHT = 20, 20  # Change the width and height to adjust the canvas size
    PIXEL_SIZE = 20  # Change the pixel size for smaller/larger pixels

    root = tk.Tk()
    root.title("Pixel Art Editor")

    editor = PixelArtEditor(root, WIDTH, HEIGHT, PIXEL_SIZE)

    colors = ['black', 'red', 'blue', 'green', 'yellow', 'orange', 'purple', 'white', 'pink', 'cyan', 'brown', 'gray']
    color_buttons = []
    for color in colors:
        color_button = tk.Button(root, bg=color, width=3, command=lambda c=color: editor.set_draw_color(c))
        color_button.pack(side='left', padx=2)
        color_buttons.append(color_button)

    clear_button = tk.Button(root, text='Clear', command=editor.clear_canvas)
    clear_button.pack()

    save_button = tk.Button(root, text='Save as Text', command=lambda: editor.save_as_text_file('pixel_art.txt'))
    save_button.pack()

    load_button = tk.Button(root, text='Load from Text', command=lambda: editor.load_from_text_file('pixel_art.txt'))
    load_button.pack()

    root.mainloop()
```

```python
import tkinter as tk
from time import strftime

def time():
    string = strftime('%H:%M:%S %p')
    label.config(text=string)
    label.after(1000, time)

def change_font_size(size):
    label.config(font=('calibri', size, 'bold'))

def change_bg_color():
    color = tk.colorchooser.askcolor(title="Choose Background Color")
    if color[1]:  # Check if a color is chosen
        label.config(bg=color[1])
        bg_color_label.config(text=f"Background Color: {color[1]}")

def change_text_color():
    color = tk.colorchooser.askcolor(title="Choose Text Color")
    if color[1]:  # Check if a color is chosen
        label.config(fg=color[1])
        text_color_label.config(text=f"Text Color: {color[1]}")

root = tk.Tk()
root.title("Digital Clock")
root.geometry("350x200")

label = tk.Label(root, font=('calibri', 40, 'bold'), background='black', foreground='white')
label.pack(anchor='center')

time()

font_size_button = tk.Button(root, text='Change Font Size', command=lambda: change_font_size(50))
font_size_button.pack(pady=10)

bg_color_button = tk.Button(root, text='Change Background Color', command=change_bg_color)
bg_color_button.pack(pady=5)

text_color_button = tk.Button(root, text='Change Text Color', command=change_text_color)
text_color_button.pack(pady=5)

# Additional labels to display the current text and background colors
text_color_label = tk.Label(root, text="Text Color: white")
text_color_label.pack(pady=5)

bg_color_label = tk.Label(root, text="Background Color: black")
bg_color_label.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk
import calendar
from datetime import datetime

def show_calendar():
    year = current_year.get()
    month = current_month.get()
    cal_content.config(text=calendar.month(year, month))

def prev_month():
    current_month.set(current_month.get() - 1)
    if current_month.get() < 1:
        current_month.set(12)
        current_year.set(current_year.get() - 1)
    show_calendar()

def next_month():
    current_month.set(current_month.get() + 1)
    if current_month.get() > 12:
        current_month.set(1)
        current_year.set(current_year.get() + 1)
    show_calendar()

# Create the Tkinter GUI
root = tk.Tk()
root.title("Calendar Application")

# Get the current year and month
current_year = tk.IntVar(value=datetime.today().year)
current_month = tk.IntVar(value=datetime.today().month)

# Create the navigation buttons
prev_button = tk.Button(root, text="Previous", command=prev_month)
prev_button.pack(side=tk.LEFT)

next_button = tk.Button(root, text="Next", command=next_month)
next_button.pack(side=tk.RIGHT)

# Create a label to display the calendar content
cal_content = tk.Label(root, text="", font=("Arial", 14))
cal_content.pack()

# Initial display of the calendar
show_calendar()

root.mainloop()
```

```python
import tkinter as tk
import random
import time

data = []

def bubble_sort():
    global data
    for i in range(len(data) - 1):
        for j in range(len(data) - 1 - i):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                draw_data(data, ['green' if x == j or x == j + 1 else 'red' for x in range(len(data))])
                time.sleep(0.1)
    draw_data(data, ['green' for x in range(len(data))])

def selection_sort():
    global data
    for i in range(len(data)):
        min_idx = i
        for j in range(i+1, len(data)):
            if data[j] < data[min_idx]:
                min_idx = j
        data[i], data[min_idx] = data[min_idx], data[i]
        draw_data(data, ['green' if x == i or x == min_idx else 'red' for x in range(len(data))])
        time.sleep(0.1)
    draw_data(data, ['green' for x in range(len(data))])

def draw_data(data, colors):
    canvas.delete("all")
    canvas_width = 800
    canvas_height = 400
    bar_width = canvas_width / len(data)

    for i, height in enumerate(data):
        x0 = i * bar_width
        y0 = canvas_height - height
        x1 = (i + 1) * bar_width
        y1 = canvas_height
        canvas.create_rectangle(x0, y0, x1, y1, fill=colors[i])
    canvas.update()

def generate_random_data():
    global data
    data = [random.randint(5, 300) for _ in range(50)]
    draw_data(data, ['red' for x in range(len(data))])

def generate_user_data():
    global data
    try:
        user_input = entry.get()
        data = [int(x) for x in user_input.split()]
        draw_data(data, ['red' for x in range(len(data))])
    except ValueError:
        pass

# Create the main application window
app = tk.Tk()
app.title("Sorting Algorithm Visualizer")

# Canvas to display bars representing data
canvas = tk.Canvas(app, width=800, height=400)
canvas.pack()

# Entry for user input
entry = tk.Entry(app)
entry.pack()

# Buttons
generate_button = tk.Button(app, text="Generate Random Data", command=generate_random_data)
generate_button.pack(side=tk.LEFT)

generate_user_button = tk.Button(app, text="Generate User Data", command=generate_user_data)
generate_user_button.pack(side=tk.LEFT)

bubble_sort_button = tk.Button(app, text="Bubble Sort", command=bubble_sort)
bubble_sort_button.pack(side=tk.LEFT)

selection_sort_button = tk.Button(app, text="Selection Sort", command=selection_sort)
selection_sort_button.pack(side=tk.LEFT)

# Generate initial random data
generate_random_data()

# Start the main event loop
app.mainloop()
```
